```python
from datasets import load_dataset

# Load labeled PAWS dataset (65k pairs)
dataset = load_dataset("paws", "labeled_final")
train_data = dataset['train']
```

```python
train_data.features
```

```python
from sentence_transformers import InputExample

# Convert PAWS sentence pairs into SBERT input format
train_samples = []
for item in train_data:
    s1, s2 = item['sentence1'], item['sentence2']
    label = float(item['label'])  # 1 = paraphrase, 0 = not
    train_samples.append(InputExample(texts=[s1, s2], label=label))

print(f"Total training samples: {len(train_samples)}")
```

```python
from sentence_transformers import SentenceTransformer

# You can switch to other models if needed (e.g., 'bert-base-nli-mean-tokens')
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
from torch.utils.data import DataLoader
from sentence_transformers import losses

# DataLoader
train_dataloader = DataLoader(train_samples, shuffle=True, batch_size=32)

# Use cosine similarity loss for training
train_loss = losses.CosineSimilarityLoss(model=model)
```

```python
# Fine-tune for 1–3 epochs (you can increase for better accuracy)
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=30,
    warmup_steps=100,
    show_progress_bar=True
)
```

```python
model_save_path = "sbert-paws-finetuned"
model.save(model_save_path)
print(f"Model saved to: {model_save_path}")
```

```python
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
from datasets import load_dataset
import numpy as np

# Load fine-tuned model
model = SentenceTransformer("sbert-paws-finetuned")  # replace with your checkpoint folder if different

# Load test data (or use validation set from PAWS)
dataset = load_dataset("paws", "labeled_final")
test_data = dataset['test']

# Prepare sentence pairs and labels
sentences1 = test_data['sentence1']
sentences2 = test_data['sentence2']
labels = test_data['label']  # 1 = paraphrase, 0 = not

# Compute embeddings and similarities
embeddings1 = model.encode(sentences1, convert_to_tensor=True, show_progress_bar=True)
embeddings2 = model.encode(sentences2, convert_to_tensor=True, show_progress_bar=True)

cosine_scores = util.cos_sim(embeddings1, embeddings2).cpu().numpy()
predicted = (cosine_scores.diagonal() >= 0.75).astype(int)  # You can tune 0.75 threshold

# Compute metrics
accuracy = accuracy_score(labels, predicted)
precision = precision_score(labels, predicted)
recall = recall_score(labels, predicted)
f1 = f1_score(labels, predicted)
conf_matrix = confusion_matrix(labels, predicted)

print(f"✅ Accuracy: {accuracy:.4f}")
print(f"✅ Precision: {precision:.4f}")
print(f"✅ Recall: {recall:.4f}")
print(f"✅ F1 Score: {f1:.4f}")
print("🧩 Confusion Matrix:\n", conf_matrix)
```
